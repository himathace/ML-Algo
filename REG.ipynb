### Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

### improt data from csv

```python
data=pd.read_csv("student_scores.csv")
data
```

### draw diagram using matplotlib 

```python
plt.scatter(data.hours_studied,data.exam_score)
plt.xlabel("hours")
plt.ylabel("marks")
```

### create a numpy array for x axis values

```python
x=np.array(data.hours_studied.values)
x
```

### create a numpy array for x axis valus

```python
y=np.array(data.exam_score.values)
y
```

### Train Model

```python
model=LinearRegression()
model.fit(x.reshape((-1,1)),y)
```

### Add new value for x

```python
new=np.array([6.2]).reshape((-1,1))
new
```

```python
predect=model.predict(new)
predect
```
